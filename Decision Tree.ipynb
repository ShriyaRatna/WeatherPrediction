```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
predictors = ["Precip", "Temp_max", "Temp_min"]
```

```python
core_weather["Main_Target"] = core_weather["Target"]//10
n = len(pd.unique(core_weather["Main_Target"]))
n
```

```python
core_weather["Main_Target"].max()
core_weather["Main_Target"].min()
```

```python
bins = [3, 6, 9, 12] 
labels = [0, 1, 2]
core_weather["Main_Target"] = pd.cut(core_weather["Target"] // 10, bins=bins, labels=labels, right=True, include_lowest=True)
```

```python
train = core_weather.loc[:"2022-12-31"]
train
```

```python
test = core_weather.loc[:"2023-01-01",:]
```

```python
test
```

```python
X_train, X_test, y_train, y_test = train_test_split(core_weather[predictors], core_weather["Main_Target"], test_size=0.2, random_state=2)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(y_test, y_pred)
```
